# Introduction
The tweet archive of Twitter user dog_rates, also known as WeRateDogs was collected in addition to tweet image prediction.
Exploratory data analysis was then performed on the cleaned dataset to give answers to the following question on the dataset:

- How does the rating of a dog image tweet affect the retweet count.
- How does the rating impact the favorite count.
- what is the tweet time distribution like and what are the effect.
- Mean confidence level of the prediction model for prediction 1 , 2, 3
- false prediction count compared to true

To answer the above questions the twitter_archive_master Dataset and tweet_image_prediction dataset was used.

# Insight 
From the analysis carried out, the following was discovered:
- From the archive tweet data, WeRateDogs had the highest number of tweets in 2016. Below is the chart showing the tweet count from the archive table by year. 

- From the archive, more tweets are recorded from weratedogs in the month of december.

- Higher number of Tweets with number of retweets above average we gotten by WeRateDogs in 2016.
- Tweets with number of retweets above average were recorded more in the month of July.

- The average prediction confidence for the predicitive model used was `59.454826%` for `prediction 1`, `13.458861%` for `prediciton 2` and `6.032417%` for `prediciton 3`
- Prediction 2 model had the highest number of correct prediction and the lowest number of wrong predictions.
- The numerator rating of a tweet does not have a correlation on the retweet count. So, the rating may not impact the retweet count

- The numerator rating of a tweet does not have a correlation on the favorite count. So, the rating may not impact the favorite count.


# Conclusion
In conclusion, from the analysis the predictive model 2 will be good for predicting tweet with dog's image and much traction should be expected from weRateDog tweeter handle in the month of December.
